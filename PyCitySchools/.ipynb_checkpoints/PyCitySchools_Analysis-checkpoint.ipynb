# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#school_data_complete.head()
#print(school_data_complete).head()
school_data_complete["school_name"].unique()
Total_Schools = len(school_data_complete["school_name"].unique())
print("Total Schools:", Total_Schools)
###############################################
total_number_students = len(school_data_complete["Student ID"])
print("Total Number of Student:", total_number_students)
###############################################
#school_data_complete.describe()
#school_data_complete.head()
#24649428 = total budget
school_data["budget"].sum()
print("Total Budget: ", school_data["budget"].sum())
################################################  how to round. can't round float
avg_read = school_data_complete["reading_score"].mean()
#avg_read.round()
#avg_read_float = float(avg_read)
#print(avg_read_float.round(2))
print("Average Reading Score: ", school_data_complete["reading_score"].mean())
################################################
avg_math = school_data_complete["math_score"].mean()
print("Average Math Score: ", school_data_complete["math_score"].mean())
################################################
overall_avg_score = ((avg_read + avg_math)/2)
print("Overall Average Score:", overall_avg_score)
################################################
num_math_passes = school_data_complete.loc[school_data_complete["math_score"] >= 70] ##
#print(num_math_passes["math_score"].count())  ##29370
count_num_math_passes = num_math_passes["math_score"].count()
percent_math_passing = ((count_num_math_passes)/(total_number_students))*100
print("Percentage of Students with a Passing Math Score:", percent_math_passing,"%")
#################################################
num_read_passes = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
#print(num_read_passes["reading_score"].count())   ##33610
count_num_read_passes = num_read_passes["reading_score"].count()
percent_read_passing = ((count_num_read_passes)/(total_number_students))*100
print("Percentage of Students with a Passing Reading Score:", percent_read_passing,"%")
```

```python
school_data
renamed = school_data.rename(columns={"school_name": "School Name", "type": "Type", "budget": "Budget", "size": "Number of Students"})
renamed
per_student_budget = renamed["Budget"]/renamed["Number of Students"]
per_student_budget
renamed['per_student_budget'] = per_student_budget
renamed
renamed2 = renamed.rename(columns={"per_student_budget": "Per Student Budget"})

renamed

```

```python
avg_math_score_per_school = school_data_complete.groupby(["school_name"])  #groupby/selects data from school_data_complete by school name
avg_math_per_school = avg_math_score_per_school["math_score"].mean()  ##takes that school name by group and finds the mean math score per school
avg_math_per_school_df = pd.DataFrame(avg_math_per_school)
avg_math_per_school_df  #creates dataframe of 1 column: #school_name is the index & math_score is the 1st column
#avg_math_per_school_df.columns #Index(['math_score'], dtype='object')
#avg_math_per_school_df
```

```python
avg_math_index = avg_math_per_school_df.reset_index()  ####resets school_name from index to 1st column and assigns new #s to index
avg_math_index ##two columns: 1. school_name 2. math_score, index from 0 - 14
avg_math_index.columns  #Index(['school_name', 'math_score'], dtype='object')
avg_math_index_renamed = avg_math_index.rename(columns={"school_name": "School Name", "math_score": "Average Math Score"}) ## renames columns
avg_math_index_renamed.columns  #Index(['School Name', 'Math Score'], dtype='object')
avg_math_index_renamed


```

```python
merge_table = pd.merge(avg_math_index_renamed, renamed2, on="School Name")
merge_table
```

```python
########### AVERAGE READING SCORE PER SCHOOL ####################
avg_read_score_per_school = school_data_complete.groupby(["school_name"])
avg_read_per_school = avg_read_score_per_school["reading_score"].mean()
avg_read_per_school_df = pd.DataFrame(avg_read_per_school)
avg_read_per_school_df#creates dataframe of 1 column: #school_name is the index & math_score is the 1st column
#avg_math_per_school_df.columns #Index(['reading_score'], dtype='object')
avg_read_index = avg_read_per_school_df.reset_index()  ####resets school_name from index to 1st column and assigns new #s to index
avg_read_index ##two columns: 1. school_name 2. reading_score, index from 0 - 14
avg_read_index.columns  #Index(['school_name', 'reading_score'], dtype='object')
avg_read_index_renamed = avg_read_index.rename(columns={"school_name": "School Name", "reading_score": "Average Reading Score"}) ## renames columns
avg_read_index_renamed.columns  #Index(['School Name', 'Reading Score'], dtype='object')
avg_read_index_renamed
############################
merge_table2 = pd.merge(avg_read_index_renamed, merge_table, on="School Name")
merge_table2
```

```python
############## PERCENT PASSING MATH BY SCHOOL   ################

#Group by school
grouped_school_data = school_data_complete.set_index('school_name').groupby(['school_name'])
#grouped_school_data.head()
# number of students per school
students_per_sch = grouped_school_data['Student ID'].count()
#students_per_sch

percent_pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_sch *100
percent_pass_math_index = percent_pass_math.reset_index()  ##reset index; now two columns, 1st column "school_name"
percent_pass_math_index_renamed = percent_pass_math_index.rename(columns={"school_name": "School Name", "Student ID": "Percent Passing Math"}) #renamed 1st column to School Name
percent_pass_math_index_renamed
merge_table3 = pd.merge(percent_pass_math_index_renamed, merge_table2, on="School Name")
merge_table3
################ % PASS READING ######################
percent_pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_sch *100
percent_pass_read_index = percent_pass_read.reset_index()  ##reset index; now two columns, 1st column "school_name"
percent_pass_read_index_renamed = percent_pass_read_index.rename(columns={"school_name": "School Name", "Student ID": "Percent Passing Reading"}) #renamed 1st column to School Name
percent_pass_read_index_renamed
merge_table4 = pd.merge(percent_pass_read_index_renamed, merge_table3, on="School Name")
merge_table4
################# OVERALL PASS RATE ####################
overall_pass_rate = (merge_table4["Percent Passing Reading"] + merge_table4["Percent Passing Math"])/2
overall_pass_rate_df = pd.DataFrame(overall_pass_rate)
#overall_pass_rate_df.columns = ["label"]
overall_pass_rate_df

merge_table5 = merge_table4.join(overall_pass_rate_df, lsuffix=" ", how="outer")
merge_table5.columns
merge_table6 =  merge_table5.rename(columns={ 0: "Overall Passing Rate"})  #index  0 no quotes, dtype float64
merge_table6

```

```python
school_summary = pd.DataFrame({
    "School Name": merge_table6["School Name"],
    "School Type": merge_table6["Type"],
    #"School ID": merge_table6["School ID"],
    "Number of Students": merge_table6["Number of Students"],
    "Budget": merge_table6["Budget"],
    "Budget per Student": merge_table6["Per Student Budget"],
    "Average Math Score": merge_table6["Average Math Score"],
    "Average Reading Score": merge_table6["Average Reading Score"],
    "Percent Passing Math": merge_table6["Percent Passing Math"],
    "Percent Passing Reading": merge_table6["Percent Passing Reading"],
    "Overall Passing Rate": merge_table6["Overall Passing Rate"]
    })

school_summary.head()
```

## Math Scores by Grade

```python
grade9_avgmathscore = pd.Series(school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name")["math_score"].mean())
grade10_avgmathscore = pd.Series(school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name")["math_score"].mean())
grade11_avgmathscore = pd.Series(school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name")["math_score"].mean())
grade12_avgmathscore = pd.Series(school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name")["math_score"].mean())

math_scores_by_grade = pd.DataFrame({
        "9th": grade9_avgmathscore,
        "10th": grade10_avgmathscore,
        "11th": grade11_avgmathscore,
        "12th": grade12_avgmathscore
})
math_scores_by_grade = math_scores_by_grade[["9th", "10th", "11th", "12th"]]
math_scores_by_grade.index.name = "School Names"
math_scores_by_grade
```
